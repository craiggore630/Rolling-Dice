```python
import numpy as np
import matplotlib.pyplot as plt

max_num_dice = int(input("Maximum number of dice to roll:  "))

rolls = [1, 2, 3, 4, 5, 6]
freqs = np.ones(6)
probs = freqs / 6
fig, axs = plt.subplots(max_num_dice // 2, 2, sharey=True, tight_layout=True)

axs[0, 0].bar(x = rolls, height = probs)

for num_dice in range(2, max_num_dice + 1):    
    new_freqs = np.zeros(num_dice * 5 + 1)
    for j in range(6):
        for k in range(len(freqs)):
            new_freqs[k+j] += freqs[k]
    freqs = new_freqs
    probs = freqs / (6 ** num_dice)
    axs[(num_dice - 1) // 2, (num_dice - 1 ) % 2].bar(x = range(num_dice, num_dice * 6 + 1), height = probs)

plt.savefig('Rolling_Dice.png')                
plt.show()
```
